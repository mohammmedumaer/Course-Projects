# 1A: English Dictionary App

This project uses the skills learnt from week 1 to week 6. The aim of this project is to create an English Dictionary app that returns definitions of English words.

```python
import json
```

```python
print("Mainmenu")
print("\n1) Add a New Word.\n2) Find the Meaning.\n3) Update a word.\n4) Exit")

while True:
    try:
        option = int(input("Select an option from given Mainmenu(1 or 2 or 3 or 4): "))
    except ValueError:
        print("Select from the given options only!!!")
    else:
        if option == 4:
            msg = input("Do you want to exit? (y/n): ")
            msg.lower()
            if msg == 'y':
                print("\nThank You, Have a Great day!")
                print("*"*60)
                break
            else:
                continue
        elif option == 1:
            word = input("Enter a word: ")
            meaning = input("Enter the meaning of the entered word: ")
            store = {word:meaning}
            with open('words.txt','a') as f:
                f.write(json.dumps(store))
                f.write("\n")
            print("Word stored sucessfully!")
            print("*"*60)
        elif option == 2:
            find = input("Enter a word to find its meaning: ")
            with open('words.txt','r') as f:
                data = f.read()
                if find in data:
                    print("Meaning of ",find,"is",meaning)
                    print('*'*60)
                    continue
                else :
                    print("Word not found")
                    print('*'*60)
        elif option == 3:
            word = input("Enter a word to be updated: ")
            with open('words.txt','r')as f:
                data = f.read()
                if word in data:
                    meaning = input("Enter new meaning of the word: ")
                    store = {word:meaning}
                    with open('words.txt','w') as f:
                        f.write(json.dumps(store))
                    print("Meaning is sucessfully updated to",meaning)
                    print('*'*60)
                    continue
                else:
                    print("Word not found")
        else:
            print("Choose correct option from the Mainmenu!!!")
```

# 1B: Library Book Management System

In this project, a Library Book Management System will be implemented with the front end GUI developed using Tkinter and the backend database implemented in SQLite. It is a single window tkinter GUI application which will implement CRUD( create, read, update, delete) operations on a ‘books’ database, in SQLite. The window will contain text widgets to accept / display book title, author, year of publication, ISBN (optional) and a large text area/ list box to display the records which match the search criteria.

```python
import sqlite3 
import tkinter as tk
import tkinter.ttk as ttk
import tkinter.messagebox as mb
import tkinter.simpledialog as sd
from tkinter import *


root = Tk()
root.title(".....Library Management Application.....")

def database():
    global conn, curr
    conn = sqlite3.connect('books.db')
    curr = conn.cursor()
    curr.execute('CREATE TABLE IF NOT EXISTS library(Bookname TEXT,Author TEXT,Year TEXT,ISBN TEXT)')
    
def create():
    if BOOKNAME.get()=="" or AUTHOR.get()=="" or YEAR.get() =="" or ISBN.get()=="":
        txt_result.config(text='Please Enter all the fields',fd='red')
    else:
        database()
        curr.execute('INSERT INTO library(Bookname,Author,Year,ISBN)VALUES(?,?,?,?)',(str(BOOKNAME.get()),str(AUTHOR.get()),str(YEAR.get()),str(ISBN.get())))
        conn.commit()
        BOOKNAME.set("")
        AUTHOR.set("")
        YEAR.set("") 
        #ISBN.set("")
        curr.close()
        conn.close()
        txt_result.config(text='Record Inserted !',fg='green')
    
def Read():
    tree.delete(*tree.get_children())
    database()
    curr.execute('SELECT * FROM library ORDER BY Author ASC ')
    fetch=curr.fetchall()
    for data in fetch:
        tree.insert('','end',values=(data[0],data[1],data[2],data[3]))
        curr.close()
        txt_result.config(text='Data Fetched Successfully!',fg='black')
        
def Exit():
    result=mb.askquestion('Do you want to Exit ?(y/n)',icon='warning')
    result.lower()
    if result == 'yes':
        root.destroy()
        exit()
        
BOOKNAME=StringVar()
AUTHOR=StringVar()
YEAR=StringVar() 
ISBN=StringVar()

Top_Frame=Frame(root,width=900,height=50,bd=8,relief='raise')
Top_Frame.pack(side=TOP)
Left_Frame=Frame(root,width=300,height=500,bd=8,relief='raise')
Left_Frame.pack(side=LEFT)
Right_Frame=Frame(root,width=600,height=500,bd=8,relief='raise')
Right_Frame.pack(side=RIGHT)

Forms=Frame(Left_Frame,width=400,height=1000)
Forms.pack(side="top")

Buttons=Frame(Left_Frame,width=300,height=100,bd=8,relief='raise')
Buttons.pack(side="bottom")

txt_title=Label(Top_Frame,width=900,font=('arial',24),text='.....Library Management Application.....')
txt_title.pack()
txt_Bookname=Label(Forms,text='Bookname',font=('arial',16),bd=15)
txt_Bookname.grid(row=0,stick='e')
txt_Author=Label(Forms,text='Author',font=('arial',16),bd=15)
txt_Author.grid(row=1,stick='e')
txt_Year=Label(Forms,text='Year',font=('arial',16),bd=15)
txt_Year.grid(row=2,stick='e')
txt_ISBN=Label(Forms,text='ISBN',font=('arial',16),bd=15)
txt_ISBN.grid(row=3,stick='e')
txt_result=Label(Buttons)
txt_result.pack(side=TOP)

Bookname=Entry(Forms,textvariable=BOOKNAME,width=30)
Bookname.grid(row=0,column=1)
Author=Entry(Forms,textvariable=AUTHOR,width=30)
Author.grid(row=1,column=1)
Year=Entry(Forms,textvariable=YEAR,width=30)
Year.grid(row=2,column=1)
ISBN=Entry(Forms,textvariable=ISBN,width=30)
ISBN.grid(row=3,column=1)

btn_display=Button(Buttons,width=10,text="Display")
btn_display.pack(side="left")
btn_search=Button(Buttons,width=10,text="Search")
btn_search.pack(side="left")
btn_add=Button(Buttons,width=10,text="Add",command=create)
btn_add.pack(side="left")
btn_issue=Button(Buttons,width=10,text="Issue",command=Read)
btn_issue.pack(side="left")                     
btn_delete=Button(Buttons,width=10,text="Delete")
btn_delete.pack(side="left")                     
btn_exit=Button(Buttons,width=10,text="Exit",command=Exit)
btn_exit.pack(side="left")      

scrollbary=Scrollbar(Right_Frame,orient=VERTICAL)
scrollbarx=Scrollbar(Left_Frame,orient=HORIZONTAL)

tree=ttk.Treeview(Right_Frame,columns=('Book','Author','Year','Isbn'))

scrollbary.config(command=tree.yview)
scrollbary.pack(side="right",fill=Y)
scrollbarx.config(command=tree.xview)
scrollbarx.pack(side="bottom",fill=X)
tree.heading('Book',text='Book',anchor=W)
tree.heading('Author',text='Author',anchor=W)
tree.heading('Year',text='Year',anchor=W)
tree.heading('Isbn',text='ISBN',anchor=W)
tree.column('#0',stretch=NO,minwidth=0,width=0)
tree.column('#1',stretch=NO,minwidth=0,width=80)
tree.column('#2',stretch=NO,minwidth=0,width=120)
tree.column('#3',stretch=NO,minwidth=0,width=160)
tree.pack()

root.mainloop()
```
